```python
import sympy
# Use p as a symbolic variable
from sympy.abc import p
# The binomial coefficient function (n choose k)
from scipy.special import comb
import math

def f(num):
    return math.factorial(num)

def get_probability_0(k_range,n,p):
    result = 1 - p
    sums = 0
    for k in range(k_range[0],k_range[1]+1):
        comb = f(n)/(f(k)*f(n-k))
        comb *= (p**k)*(1-p)**(8-k)
        sums += comb
    return result * sums

def get_probability_1(k_range,n,p):
    result = p
    sums = 0
    for k in range(k_range[0],k_range[1]+1):
        comb = f(n)/(f(k)*f(n-k))
        comb *= (p**k)*(1-p)**(8-k)
        sums += comb
    return result * sums

from sympy.abc import p
a = get_probability_0((0,4),8,p)
b = get_probability_0((8,5),8,p)
c = get_probability_1((0,3),8,p)
d = get_probability_1((4,8),8,p)

#integrate to 1
print(sympy.simplify(a+b+c+d).subs(p,0))
```

```python
import numpy as np
import matplotlib.pyplot as plt


def plot_cobweb(f, x0, nmax=20):
    x = np.linspace(0, 1, 50)
    fig = plt.figure()
    ax = fig.add_subplot(111)

    ax.plot(x, f(x))
    ax.plot(x, x)

    px, py = np.empty((2,nmax+1,2))
    px[0], py[0] = x0, 0
    for n in range(1, nmax, 2):
        px[n] = px[n-1]
        py[n] = f(px[n-1])
        px[n+1] = py[n]
        py[n+1] = py[n]

    ax.plot(px, py)
```

```python
def func(x):
    return x**3*(x-1)**5*(28*x-84)

plot_cobweb(func, 0.5)
plot_cobweb(func, 0.15)

#stationary points at 0.4 (stable) and 0.2 (unstable)
```
