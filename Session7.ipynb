```python
import sympy
# Use p as a symbolic variable
from sympy.abc import p
# The binomial coefficient function (n choose k)
from scipy.special import comb

result = ((1-p) * comb(8, 3) * p**3 * (1-p)**5 + p * sum([comb(8, k) * p**k * (1-p)**(8-k) for k in range(2, 4)]))
print(sympy.simplify(result))
```

```python
import numpy as np
import matplotlib.pyplot as plt


def plot_cobweb(f, x0, nmax=20):
    x = np.linspace(0, 1, 50)
    fig = plt.figure()
    ax = fig.add_subplot(111)

    ax.plot(x, f(x))
    ax.plot(x, x)

    px, py = np.empty((2,nmax+1,2))
    px[0], py[0] = x0, 0
    for n in range(1, nmax, 2):
        px[n] = px[n-1]
        py[n] = f(px[n-1])
        px[n+1] = py[n]
        py[n+1] = py[n]

    ax.plot(px, py)
```

```python
def func(x):
    return x**3*(x-1)**5*(28*x-84)

plot_cobweb(func, 0.5)
plot_cobweb(func, 0.9)
```
