```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

```python
def f(x):
    Z = 24.44321494051954
    if abs(x) > 7:
        return 0
    elif abs(x) > 3:
        return 3 * (1-(x / 7) ** 2) ** 0.5 / Z
    elif abs(x) > 1:
        return ((3- abs(x))/2 - 3/7 * 10**0.5 * ((3 - x**2 + 2*abs(x))**0.5-2)) / Z
    elif abs(x) > 0.75:
        return (9-8 * abs(x)) / Z
    elif abs(x) > 0.5:
        return (3 * abs(x) + 0.75) / Z
    else:
        return 2.25 / Z
```

```python
def simulation(f, x0, steps):
    states = []
    x = x0
    
    for i in range(steps):
        y = np.random.normal(x, 2)
        if np.random.uniform() < min(1, f(y) / f(x)):
            x = y
        if i % 100 == 0:
            states.append(x)
            
    return states

x = np.arange(-8, 8, 0.1)
y = [f(xi) for xi in x]

plt.figure()
mh = simulation(f, x0 = 0, steps=100000)
plt.plot(x, y)
plt.hist(mh, density=True, bins = 50)
plt.show()
```
